Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Data collection and processing

```python
sonar_data = pd.read_csv('sonardata.csv', header=None) # header= None means in a file there is not any label/name of row and column

sonar_data.head()
```

```python
# no of rows and columns

sonar_data.shape
```

```python
sonar_data.describe() #describe gives statistical measures of the data
```

```python
sonar_data[60].value_counts() #this is not sufficient for train a model in machine learning 

# note that more the data more accurate the model
```

```python
sonar_data.groupby(60).mean()

# here i divided the rock and mine into 2 groups and calculate the mean of each row
```

```python
# saperating data and labels basically i am using supervised learning to train the model

X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
```

```python
print(X)
```

```python
print(Y)
```

#training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.1,stratify=Y, random_state=1)

# so here in paremeters test_size = 0.1 means 10 % if i wrote 0.3 that means 30% data

# stratify = Ensures that the test and train splits have the same ratio of class ratio for training classification models. We use the stratify parameter and pass the y series.
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

model training  --- > logistic regression model

```python
model = LogisticRegression()
```

```python
# training hte logistic regression model with training data

model.fit(X_train, Y_train)
```

```python
# model evaluation
```

```python
# accuracy on the training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data : ', training_data_accuracy)
```

```python
# ACCURACY ON TEST DATA
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on test data : ', test_data_accuracy)
```
